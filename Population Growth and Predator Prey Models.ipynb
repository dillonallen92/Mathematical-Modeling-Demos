# Population Growth Modeling / Predator-Prey

## First Encounter (Logistic Growth)

The first example here is a common problem seen in second semester Calculus classes with the introduction of differential equations.

The common form seen in Calculus is the logistic model

$$ \frac{dy}{dt} = ry \left( 1 - \frac{y}{K}\right)$$

With the solution

$$ y = \frac{K}{1 + Ae^{-rt}} $$

With constants:

- K: Carrying capacity for the population

- A: Integration constant, Considered initial population point

- r: Positive constant for growth rate

This model assumes a population *does not* have unlimited resources, but instead moves between 0 and the capacity K and treats K as an asymptote.

## Code for Logistic Models

```python
# Basic Library imports

import numpy as np
import matplotlib.pyplot as plt

# Define the following constants
K = 600
A = 200
r = 0.7
cycles = 20
ts = 1 # Time step of 1 (unit of time)

y = [1]

for i in range(0, cycles):
    new_y = y[i] + r*y[i]*(1 - y[i]/K)*ts
    y.append(new_y)

t = range(cycles + 1)
plt.plot(t, y)
plt.xlabel("Time")
plt.ylabel("Population")
plt.title("Logistic Growth with Carry Capacity 250")
```

## Predator - Prey Models

As we can see in the example above, the Logistic Growth equation does not have any sort of functionality besides showing a population grow to its capacity.

In reality, no population grows unchecked. The subjects in question will die of old age, sickness, and from predators who decide they make good meals. This leads us to the **Predator-Prey Model**

An example of this would be a population of Sheep vs Wolves:

$$ \frac{dS}{dt} =  G_s \cdot S - D_s \cdot S \cdot W $$


$$ \frac{dW}{dt} = G_w \cdot S \cdot W - D_w \cdot W $$

With the values:

- $S$ : Number of sheep at the start of the time period.
- $G_s$ : Growth rate of the sheep.
- $D_s$ : Death rate of the sheep.
- $W$ : Number of wolves at the start of the time period.
- $G_w$ : Growth rate of the wolves.
- $D_w$ : Death rate of the wolves.

This system is more complicated because they are coupled differential equations. Each variable relies on the other to predict the next value at time-step $\Delta t$. We can instead turn this into a difference equation (makes it easier to translate into code):

$$ S_{t+1} = S_t + \Delta t \cdot \left( S_t \cdot G_s - S_t \cdot D_s \cdot W_t \right) $$

$$ W_{t+1} = W_t + \Delta t \cdot \left( W_t \cdot G_w \cdot S_t - W_t \cdot D_w \right) $$

Now, lets write a simulation before for a cycle of, say, 400 years.

```python
import numpy as np
import matplotlib.pyplot as plt

#  --- Variables and conditions ---

# Time conditions
ts = 1/12 
cycles = 400

# Start an array of 400 sheep, and set initial population to 500
S = np.zeros(cycles + 1)
S[0] = 500

# Array of 400 wolves, initial population 40
W = np.zeros(cycles + 1)
W[0] = 40

# Add in biological factors
sheep_birth_rate = 0.2
sheep_death_rate = 0.003
wolf_birth_rate = 0.001
wolf_death_rate = 0.5

# --- Difference Equation ---

for i in range(cycles):
    S[i+1] = S[i] + ts * ( S[i] * sheep_birth_rate - S[i]* sheep_death_rate * W[i])
    W[i+1] = W[i] + ts * ( W[i] * wolf_birth_rate * S[i] - W[i] * wolf_death_rate)

# --- Plotting ---

t = range(cycles + 1)

fig = plt.figure()

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(t,S, 'g-')
ax1.set_ylabel('Sheep Population', color='green')
for t1 in ax1.get_yticklabels():
    t1.set_color('g')

ax2 = ax1.twinx()
ax2.plot(t, W, 'r-')
ax2.set_ylabel('Wolves', color='red')
for tl in ax2.get_yticklabels():
    tl.set_color('r')


plt.show()
```
